## Subselections

```r
# First lines
head(pokemon.df)
```

```r
# Selecting rows
pokemon.df[10:15, ]
```

```r
# Selecting one column (A list)
pokemon.df$"Attack"
```

```r
# Selecting more than one column
head(pokemon.df[, c("Attack", "Defense", "Speed")])
```

```r
# Removing rows
head(pokemon.df[-c(1,2,3), ])  # Python starts on index 0, on R we starts on 1
```

```r
# Removing columns
head(subset(pokemon.df, select=-c(Type.2, Generation)))
```

```r
# Filtering
pokemon.df[pokemon.df$Attack>180, ]
```

```r
# Sorting by attack
head(pokemon.df[order(pokemon.df$Attack), ])
```

## Operations between columns

```r
# Functions between columns
pokemon.df$Power <- pokemon.df$Attack - pokemon.df$Defense
head(pokemon.df)
```

```r
pokemon.df$Attack2 <- pokemon.df$Attack * 2
head(pokemon.df)
```
